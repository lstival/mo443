```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Read the image

```python
#Read image in BGR
image = cv2.imread('baboon.png')
plt.imshow(image)
```

```python
#Convert to RGB
im_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
#Convert to HSV
img_hsv = cv2.cvtColor(im_rgb, cv2.COLOR_RGB2HSV)
```
