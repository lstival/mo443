```python
import numpy as np
from skimage import io, color
```

## Reading the image

```python
img = io.imread("baboon.png")
io.imshow(img) 
io.show()
```

```python
img.shape
```

### Converting to CIELab

```python
lab = color.rgb2lab(img)
lab.shape
```

```python
#Define the threshold value to define the 0 or 1
threshold = lab[:,:,0].max() / 2
```

## Appling the mask

```python
#Set the luminance channel to be used
luminance = lab[:,:,2]

#Define the matriz with erro of each pixel
erro = np.zeros([luminance.shape[0], luminance.shape[1]])

#Output image
g = luminance.copy()
```

```python
# lab2 = lab.copy()
```

```python
def create_middle_tons(luminance, g, erro, mask, limit=1):
    """
    Loop to apply the middle tons mask in each pixel of the image
    """
    #loop to check all pixels of the image
    for row in range(luminance.shape[0]-limit):
        for collumn in range(luminance.shape[1]-limit):
            if luminance[row][collumn] < threshold:
                g[row][collumn] = 0
            else:
                g[row][collumn] = 1

            erro[row][collumn] = luminance[row][collumn] - g[row][collumn] * 255

            #Applyig the mask
            luminance, erro = mask(luminance, erro, row, collumn)
    return luminance
```

## Mask functions

```python
def mask_Steinberg(luminance, erro, row, collumn):
    """
    Apply the Floyd e Steinberg in a pixel of the image, and return the luminance and erro;
    """
    luminance[row+1][collumn] = luminance[row+1][collumn] + (7/16) * erro[row+1][collumn]
    luminance[row-1][collumn-1] = luminance[row-1][collumn-1] + (3/16) * erro[row-1][collumn-1]
    luminance[row][collumn+1] = luminance[row][collumn+1] + (5/16) * erro[row][collumn+1]
    luminance[row+1][collumn+1] = luminance[row+1][collumn+1] + (1/16) * erro[row+1][collumn+1]
    
    return luminance, erro 
```

```python
def mask_Stevenson(luminance, erro, row, collumn):
    """
    Apply the Stevenson in a pixel of the image, and return the luminance and erro;
    """
    #Same line
    luminance[row+2][collumn] = luminance[row+2][collumn] + (32/200) * erro[row+2][collumn]
    
    #Line -1
    luminance[row-3][collumn-1] = luminance[row-3][collumn-1] + (12/200) * erro[row-3][collumn-1]
    luminance[row-1][collumn-1] = luminance[row-1][collumn-1] + (26/200) * erro[row-1][collumn-1]
    luminance[row+1][collumn-1] = luminance[row+1][collumn-1] + (30/200) * erro[row+1][collumn-1]
    luminance[row+3][collumn-1] = luminance[row+3][collumn-1]  + (16/200) * erro[row+3][collumn-1]
    
    #Line -2
    luminance[row-2][collumn-2] = luminance[row-2][collumn-2] + (12/200) * erro[row-2][collumn-2]
    luminance[row][collumn-2] = luminance[row][collumn-2] + (26/200) * erro[row][collumn-2]
    luminance[row+2][collumn-2] = luminance[row+2][collumn-2] + (12/200) * erro[row+2][collumn-2]
    
    #Line -3
    luminance[row-3][collumn-3] = luminance[row-3][collumn-3] + (5/200) * erro[row-3][collumn-3]
    luminance[row-1][collumn-3] = luminance[row-1][collumn-3] + (12/200) * erro[row-1][collumn-3]
    luminance[row+1][collumn-3] = luminance[row+1][collumn-3] + (12/200) * erro[row+1][collumn-3]
    luminance[row+3][collumn-3] = luminance[row+3][collumn-3] + (5/200) * erro[row+3][collumn-3]
    
    return luminance, erro 
```

```python
def mask_Burkes(luminance, erro, row, collumn):
    """
    Apply the Burkes in a pixel of the image, and return the luminance and erro;
    """
    #Same line
    luminance[row+1][collumn] = luminance[row+1][collumn] + (8/32) * erro[row+1][collumn]
    luminance[row+2][collumn] = luminance[row+2][collumn] + (4/32) * erro[row+2][collumn]
    
    #Line -1
    luminance[row-2][collumn-1] = luminance[row-1][collumn-2] + (2/32) * erro[row-1][collumn-2]
    luminance[row-1][collumn-1] = luminance[row-1][collumn-1] + (4/32) * erro[row-1][collumn-1]
    luminance[row][collumn-1] = luminance[row][collumn-1] + (8/32) * erro[row][collumn-1]
    luminance[row+1][collumn-1] = luminance[row+1][collumn-1] + (4/32) * erro[row+1][collumn-1]
    luminance[row+2][collumn-1] = luminance[row+2][collumn-1] + (2/32) * erro[row+2][collumn-1]
    
    return luminance, erro 
```

```python
def mask_Sierra(luminance, erro, row, collumn):
    """
    Apply the Sierra in a pixel of the image, and return the luminance and erro;
    """
    #Same Line
    luminance[row][collumn+1] = luminance[row][collumn+1] + (5/32) * erro[row][collumn+1]
    luminance[row][collumn+2] = luminance[row][collumn+2] + (3/32) * erro[row][collumn+2]
    
    #Line -1
    luminance[row-1][collumn-2] = luminance[row-1][collumn-2] + (2/32) * erro[row-1][collumn-2]
    luminance[row-1][collumn-1] = luminance[row-1][collumn-1] + (4/32) * erro[row-1][collumn-1]
    luminance[row-1][collumn] = luminance[row-1][collumn] + (5/32) * erro[row-1][collumn]
    luminance[row-1][collumn+1] = luminance[row-1][collumn+1] + (4/32) * erro[row-1][collumn+1]
    luminance[row-1][collumn+2] = luminance[row-1][collumn+2] + (2/32) * erro[row-1][collumn+2]
    
    #Line-2
    luminance[row-2][collumn-1] = luminance[row-2][collumn-1] + (2/32) * erro[row-2][collumn-1]
    luminance[row-2][collumn] = luminance[row-2][collumn] + (3/32) * erro[row-2][collumn]
    luminance[row-2][collumn+1] = luminance[row-2][collumn+1] + (2/32) * erro[row-2][collumn+1]
    
    return luminance, erro 
```

```python
def mask_Stucki(luminance, erro, row, collumn):
    """
    Apply the Sierra in a pixel of the image, and return the luminance and erro;
    """
    #Same Line
    luminance[row][collumn+1] = luminance[row][collumn+1] + (8/42) * erro[row][collumn+1]
    luminance[row][collumn+2] = luminance[row][collumn+2] + (4/42) * erro[row][collumn+2]
    
    #Line -1
    luminance[row-1][collumn-2] = luminance[row-1][collumn-2] + (2/42) * erro[row-1][collumn-2]
    luminance[row-1][collumn-1] = luminance[row-1][collumn-1] + (4/42) * erro[row-1][collumn-1]
    luminance[row-1][collumn] = luminance[row-1][collumn] + (8/42) * erro[row-1][collumn]
    luminance[row-1][collumn+1] = luminance[row-1][collumn+1] + (4/42) * erro[row-1][collumn+1]
    luminance[row-1][collumn+2] = luminance[row-1][collumn+2] + (2/42) * erro[row-1][collumn+2]
    
    #Line-2
    luminance[row-2][collumn-2] = luminance[row-2][collumn-2] + (1/42) * erro[row-2][collumn-2]
    luminance[row-2][collumn-1] = luminance[row-2][collumn-1] + (2/42) * erro[row-2][collumn-1]
    luminance[row-2][collumn] = luminance[row-2][collumn] + (4/42) * erro[row-2][collumn]
    luminance[row-2][collumn+1] = luminance[row-2][collumn+1] + (2/42) * erro[row-2][collumn+1]
    luminance[row-2][collumn+2] = luminance[row-2][collumn+2] + (1/42) * erro[row-2][collumn+2]
    
    return luminance, erro 
```

```python
def mask_Jarvis(luminance, erro, row, collumn):
    """
    Apply the Sierra in a pixel of the image, and return the luminance and erro;
    """
    #Same Line
    luminance[row][collumn+1] = luminance[row][collumn+1] + (7/48) * erro[row][collumn+1]
    luminance[row][collumn+2] = luminance[row][collumn+2] + (5/48) * erro[row][collumn+2]
    
    #Line -1
    luminance[row-1][collumn-2] = luminance[row-1][collumn-2] + (3/48) * erro[row-1][collumn-2]
    luminance[row-1][collumn-1] = luminance[row-1][collumn-1] + (5/48) * erro[row-1][collumn-1]
    luminance[row-1][collumn] = luminance[row-1][collumn] + (7/48) * erro[row-1][collumn]
    luminance[row-1][collumn+1] = luminance[row-1][collumn+1] + (5/48) * erro[row-1][collumn+1]
    luminance[row-1][collumn+2] = luminance[row-1][collumn+2] + (3/48) * erro[row-1][collumn+2]
    
    #Line-2
    luminance[row-2][collumn-2] = luminance[row-2][collumn-2] + (1/48) * erro[row-2][collumn-2]
    luminance[row-2][collumn-1] = luminance[row-2][collumn-1] + (3/48) * erro[row-2][collumn-1]
    luminance[row-2][collumn] = luminance[row-2][collumn] + (5/48) * erro[row-2][collumn]
    luminance[row-2][collumn+1] = luminance[row-2][collumn+1] + (3/48) * erro[row-2][collumn+1]
    luminance[row-2][collumn+2] = luminance[row-2][collumn+2] + (1/48) * erro[row-2][collumn+2]
    
    return luminance, erro 
```

## Aplying the masks

```python
#Jarvis Result
lab_Jarvis = lab.copy()
luminance_Jarvis= luminance.copy()
g_Jarvis = luminance.copy()

#Define the matriz with erro of each pixel
erro = np.zeros([luminance_Jarvis.shape[0], luminance_Jarvis.shape[1]])

luminance_Jarvis = create_middle_tons(luminance_Jarvis, g_Jarvis, erro, mask_Jarvis, limit=2)
lab_Jarvis[:,:,2] = luminance_Jarvis
io.imshow(img)
io.show()
io.imshow(color.lab2rgb(lab_Jarvis)) 
io.show()
```
